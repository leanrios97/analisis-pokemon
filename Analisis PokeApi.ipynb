##### Funciones para obtener los datos.

```python
def obtenerNombre(data, count):
    try:
        idPoxedex = count + 1
        name = data['name'].capitalize()
        return idPoxedex, name
    
    except Exception as e:
        print("No se pudo obtener el nombre.")
        name = np.nan
        return name
```

```python
def obtenerTipo(data):
    try:
        tipoCant = len(data['types'])
        
        if tipoCant == 1:
            tipo1 = data['types'][0]['type']['name']
            tipo2 = np.nan
            return tipo1, tipo2
            
        if tipoCant >= 1: 
            tipo1 = data['types'][0]['type']['name']
            tipo2 = data['types'][1]['type']['name']
            return tipo1, tipo2
            
    except Exception as e:
        print("No se puede obtener el tipo")
        name = np.nan
        return name
```

```python
def obtenerGeneracion(idPokedex):
    n_dict_generation = {'1':[151,0,151],
                         '2':[100,152,251], 
                         '3':[135,252,386],
                         '4':[107, 387, 493],
                         '5':[156, 494, 649],
                         '6':[72, 650, 721],
                         '7':[88, 722, 809],
                         '8':[96, 810, 905],
                         '9':[120, 906, 1025]}
    try:
        if idPokedex > n_dict_generation['1'][1] and idPokedex <= n_dict_generation['1'][2]:
            generation = int(1)
            return generation
            
        if idPokedex > n_dict_generation['2'][1] and idPokedex <= n_dict_generation['2'][2]:
            generation = int(2)
            return generation
            
        if idPokedex > n_dict_generation['3'][1] and idPokedex <= n_dict_generation['3'][2]:
            generation = int(3)
            return generation
            
        if idPokedex > n_dict_generation['4'][1] and idPokedex <= n_dict_generation['4'][2]:
            generation = int(4)
            return generation
            
        if idPokedex > n_dict_generation['5'][1] and idPokedex <= n_dict_generation['5'][2]:
            generation = int(5)
            return generation
            
        if idPokedex > n_dict_generation['6'][1] and idPokedex <= n_dict_generation['6'][2]:
            generation = int(6)
            return generation
            
        if idPokedex > n_dict_generation['7'][1] and idPokedex <= n_dict_generation['7'][2]:
            generation = int(7)
            return generation
            
        if idPokedex > n_dict_generation['8'][1] and idPokedex <= n_dict_generation['8'][2]:
            generation = int(8)
            return generation

        if idPokedex > n_dict_generation['9'][1] and idPokedex <= n_dict_generation['9'][2]:
            generation = int(9)
            return generation
        
    except Exception as e:
        print(f"Error: {e}")
        return np.nan
```

## Armado DataFrame

```python
columns = ['IdPokedex', 'Nombre', 'Generaci√≥n','Tipo1', 'Tipo2']
```

```python
df_pokemons = pd.DataFrame(columns=columns)
df_pokemons
```

```python
idPokedex = 0
for pokemon in pokemons:
    
    response = requests.get(pokemon['url'])

    if response.status_code == 200:
        data = response.json()
        
        idPokedex, name = obtenerNombre(data, idPokedex)
        generation = obtenerGeneracion(idPokedex)
        tipo1, tipo2 = obtenerTipo(data)
        
        nueva_fila = pd.Series([idPokedex, name, generation, tipo1, tipo2], index = df_pokemons.columns)
        df_pokemons = df_pokemons.append(nueva_fila, ignore_index=True)
        
        
    else: 
        raise ValueError(f'Error al realizar la solicitud: {response.status_code}')
```

```python
df_pokemons.to_csv("df_pokemons.csv", sep=';')
```
